# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(lt):
    return list(map(lambda ele: 0 if ele<0 else ele, lt))
def dReLU(lt):
    return list(map(lambda ele: 0 if ele<=0 else 1, lt))
x = list(plt.linspace(-10,10,100))
X = ReLU(x)
dX = dReLU(X)
plt.plot(x, X, 'b')#ReLU
plt.plot(x,dX, 'r')#dReLU
plt.title('ReLU f()')
plt.grid()
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.ylabel('f(x)')
plt.xlabel('linspace(-10,10,10)')
plt.legend(['ReLU', 'dReLU'], loc='upper left')
plt.show()
```
