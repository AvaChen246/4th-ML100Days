```python
df = app_train
df.head(10)
```

```python
DF = df['AMT_INCOME_TOTAL'].describe()
DF
```

```python
x = np.array(df['SK_ID_CURR'].head())
y = np.array(df['AMT_INCOME_TOTAL'].head())
plt.bar(x,y,align='center') # A bar chart
plt.title('AMT_INCOME_TOTAL')
plt.xlabel('SK_ID_CURR')
plt.ylabel('AMT_INCOME_TOTAL')
plt.show()
#-----------------------------
print(df['AMT_INCOME_TOTAL'].describe())
plt.hist(np.array(df['AMT_INCOME_TOTAL']),bins=50)
plt.title('AMT_INCOME_TOTAL')
plt.xlim(0, 0.2*10**8)
plt.show()
print(df['AMT_CREDIT'].describe())
plt.hist(df['AMT_CREDIT'],bins=50)
plt.title('AMT_CREDIT')
plt.xlim(0, 3000000)
plt.show()
print(df['AMT_ANNUITY'].describe())
plt.hist(df['AMT_ANNUITY'],bins=50)
plt.title('AMT_ANNUITY')
plt.xlim(0, 120000)
plt.show()
#------------------------------
```
